```python
from ultralytics import YOLO
from google.colab import files
from IPython.display import Video, display
import os

print("Libraries imported successfully")
```

```python
model = YOLO("yolo11n.pt")

print("YOLO model loaded successfully")
```

```python
uploaded = files.upload()
```

```python
video_path = list(uploaded.keys())[0]

print("Uploaded video:", video_path)
```

```python
results = model.track(
    source=video_path,
    tracker="bytetrack.yaml",
    conf=0.35,
    iou=0.5,
    save=True,
    persist=True
)

print("Object detection and tracking completed")
```

```python
import glob

output_files = (
    glob.glob("runs/track/**/*.mp4", recursive=True)
    + glob.glob("runs/track/**/*.avi", recursive=True)
    + glob.glob("runs/detect/**/*.mp4", recursive=True)
    + glob.glob("runs/detect/**/*.avi", recursive=True)
)

print("Output files:", output_files)
```
